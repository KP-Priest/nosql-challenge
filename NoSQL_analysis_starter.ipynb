```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
first_document = establishments.find_one(query)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of documents with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
pprint(first_document)

```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(list(first_document))


# Display the number of rows in the DataFrame
print(f"Number of rows: {len(hygiene_df)}")

# Display the first 10 rows of the DataFrame
print(hygiene_df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": {"$regex" : "London"},
         "RatingValue": {"$gte": 4}
}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments in London with rating >= 4: {count}")


# Display the first document in the results using pprint

first_document = establishments.find_one(query)
print("First establishment in London with rating >= 4:")
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
london_high_rated_df = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(london_high_rated_df)}")


# Display the first 10 rows of the DataFrame
london_high_rated_df.head(10)

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Find coordinates of Penang Flavours
penang = establishments.find_one({"BusinessName": "Penang Flavours"})
penang_lat = float(penang['geocode']['latitude'])
penang_lon = float(penang['geocode']['longitude'])


degree_search = 0.01

# Build query
query = {
    "RatingValue": 5,
    "geocode.latitude": {
        "$gte": penang_lat - degree_search,
        "$lte": penang_lat + degree_search
    },
    "geocode.longitude": {
        "$gte": penang_lon - degree_search,
        "$lte": penang_lon + degree_search
    },
    "scores.Hygiene": {"$gte": 1}
}

# Sort by lowest hygiene score
sort = [("scores.Hygiene", 1)] 

limit = 5


# Print the results
results = establishments.find(query).sort(sort).limit(limit)
for result in results:
    pprint(result)
    
```

```python
# Convert to DataFrame
nearby_high_rated_df = pd.DataFrame(establishments.find(query).sort(sort).limit(limit))

# Display the results
print(f"Number of rows in DataFrame: {len(nearby_high_rated_df)}")
nearby_high_rated_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    
    # Group by Local Authority
    {"$group": {
        "_id": "$LocalAuthorityName",
        "count": {"$sum": 1}
    }},
    {"$sort": {"count": -1}}
]

results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(results)}")


# Print the first 10 results
print("\nFirst 10 results:")
for i, result in enumerate(results[:10], 1):
    print(f"{i}. {result['_id']}: {result['count']} establishments")
    
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(list(establishments.aggregate(pipeline)))


# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame; {len(hygiene_df)}")
print("\nFirst 10 rows of the DataFrame:")

# Display the first 10 rows of the DataFrame
print(hygiene_df.head(10))
      
```
